## Participación de Mercado

```python
#Market Share 2023
dfMarketShare = df1.loc["2023"]
```

```python
dfMarketShare = dfMarketShare[['Sistema', 'BBVA México', 'Banorte', 'Santander', 'Scotiabank', 'HSBC', 'Banamex']]
```

```python
dfMarketShare
```

```python
dfSistema = dfMarketShare['Sistema']
```

```python
dfSistema
```

```python
for column in dfMarketShare:
  dfMarketShare[column]= (dfMarketShare[column] / dfSistema)*100
```

```python
dfMarketShare
```

```python
bk1 = 'BBVA México'
bk2 = 'Banorte'
bk3 = 'Santander'
bk4 = 'Scotiabank'
bk5 = 'HSBC'
bk6 = 'Banamex'
mercado = (dfMarketShare[bk1].iloc[-1] + dfMarketShare[bk2].iloc[-1]+ dfMarketShare[bk3].iloc[-1]+
              dfMarketShare[bk4].iloc[-1]+ dfMarketShare[bk5].iloc[-1]+ dfMarketShare[bk6].iloc[-1])
bk7  = 100 - mercado
```

```python
bk7
```

```python
labels = bk1,bk2,bk3,bk4,bk5,bk6, 'Otras EF'
OtrasEF = bk7
sizes = [dfMarketShare[bk1].iloc[-1], dfMarketShare[bk2].iloc[-1], dfMarketShare[bk3].iloc[-1],
              dfMarketShare[bk4].iloc[-1], dfMarketShare[bk5].iloc[-1], dfMarketShare[bk6].iloc[-1],OtrasEF ]

fig7, ax7 = plt.subplots()
ax7.set_title('Participación de Mercado', color='saddlebrown')

ax7.pie(sizes, labels=labels, autopct='%1.1f%%')
```
