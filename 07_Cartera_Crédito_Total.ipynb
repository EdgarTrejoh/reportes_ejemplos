## Análiis de la Cartera de Consumo

```python
# Composición Cartera Consumo 2023
dfConsumo = df1.loc["2023"]
```

```python
dfConsumo = dfConsumo[['Consumo', 'Tarjeta Crédito', 'Nómina', 'Personales', 'Auto']]
```

```python
dfConsumo
```

```python
dfConsumo['subtotal'] = dfConsumo[['Tarjeta Crédito', 'Nómina', 'Personales', 'Auto']].apply(np.sum, axis=1)
```

```python
dfConsumo['Otros Créditos'] = dfConsumo['Consumo'] - dfConsumo['subtotal']
```

```python
dfConsumo.drop(['subtotal'], axis = 'columns', inplace=True)
```

```python
dfConsumo.rename(columns={"Consumo": "Cartera Total", "Nómina": "Crédito Nómina", "Personales": "Crédito Personal", "Auto": "Crédito Automotríz"}, inplace=True)
```

```python
dfConsumo
```

```python
dfComp_Cartera = dfConsumo.copy()
```

```python
df_Cartera_total = dfConsumo['Cartera Total']
```

```python
for column in dfConsumo:
  dfComp_Cartera[column] = (dfConsumo[column] / df_Cartera_total)*100
```

```python
dfComp_Cartera
```

```python
fig6, ax6 = plt.subplots()
ax6.set_title('Composición de la Cartera de Crédito al Consumo [Total] \n Cifras al mes de Julio 2023', color='saddlebrown')

label = ['Tarjeta Crédito', 'Crédito Nómina', 'Crédito Personal', 'Crédito Automotríz', 'Otros Créditos']

ax6.pie(dfConsumo[['Tarjeta Crédito', 'Crédito Nómina', 'Crédito Personal', 'Crédito Automotríz', 'Otros Créditos']].iloc[-1], labels=label, autopct='%1.1f%%')
```
